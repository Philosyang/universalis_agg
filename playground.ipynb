```python
import requests
import json
```

```python
string_query = '沙龙'
accepted_types = ['Item', 'Mount', 'Balloon', 'BuddyEquip', 'Orchestrion']  # https://xivapi.com/docs/Search#search (Accepts: Achievement, Title, Action, CraftAction, Trait, PvPAction, PvPTrait, Status, BNpcName, ENpcResident, Companion, Mount, Leve, Emote, InstanceContent, Item, Recipe, Fate, Quest, ContentFinderCondition, Balloon, BuddyEquip, Orchestrion, PlaceName, Weather, World, Map, lore_finder)
datacenter = '猫小胖'
world = '摩杜纳'
```

```python
x = requests.get(
    'https://cafemaker.wakingsands.com/search?string=' + string_query)
# print(x.text)
y = json.loads(x.text)
```

```python
y   # python dict
```

```python
results = y['Results']
results
```

```python
name_id_dict = {}

for i in results:
    if i['UrlType'] in accepted_types:
        name_id_dict[i['Name']] = i['ID']

name_id_dict
```
